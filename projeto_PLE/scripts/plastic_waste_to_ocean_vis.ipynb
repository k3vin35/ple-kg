```python
#bibliotecas
import pandas as pd
```

```python
#carregar dataset
data = pd.read_csv('C:\\uni\\ano2\\sem1\\vis_info\\projeto_PLE\\datasets\\plastic-waste-emitted-to-the-ocean.csv')
```

```python
#Infos do dataset
print(data.head())
print(data.info())
print(data.describe())
print(data.columns)
```

```python
#alterações no dataset
data.rename(columns={'Entity': 'Country'}, inplace=True)
data.rename(columns={'Mismanaged waste emitted to the ocean (metric tons year-1)': 'PlasticEmittedToOcean'}, inplace=True)
```

```python
#visualizações dos dados
data['Country'].unique()

```

```python
regions = [
    "Africa",
    "Asia",
    "Europe",
    "North America",
    "South America",
    "Oceania",
    "World",
    "EU-27"
]


# separar datasets
df_regions = data[data["Country"].isin(regions)].copy()
df_countries = data[~data["Country"].isin(regions)].copy()

# guardar CSVs
df_countries.to_csv("C:\\uni\\ano2\\sem1\\vis_info\\projeto_PLE\\datasets\\emitted_countries.csv", index=False)
df_regions.to_csv("C:\\uni\\ano2\\sem1\\vis_info\\projeto_PLE\\datasets\\emitted_regions.csv", index=False)

```
