```python
#bibliotecas
import pandas as pd
```

```python
#carregar dataset
data = pd.read_csv('plastic-waste-emitted-to-the-ocean-per-capita.csv')
```

```python
#Infos do dataset
print(data.head())
print(data.info())
print(data.describe())
print(data.columns)
```

```python
#alterações no dataset
data.rename(columns={'Entity': 'Country'}, inplace=True)
data.rename(columns={'Mismanaged plastic waste to ocean per capita (kg per year)': 'PlasticEmittedPerCapita'}, inplace=True)
```

```python
#visualizações dos dados
data['Country'].unique()

```

```python
regions = [
    "Africa",
    "Asia",
    "Europe",
    "North America",
    "Oceania",
    "World"
]



# separar datasets
df_regions = data[data["Country"].isin(regions)].copy()
df_countries = data[~data["Country"].isin(regions)].copy()

# guardar CSVs
df_countries.to_csv("percapita_countries.csv", index=False)
df_regions.to_csv("percapita_regions.csv", index=False)

```
