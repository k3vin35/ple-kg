```python
#bibliotecas
import pandas as pd
```

```python
#carregar dataset
data = pd.read_csv('datasets\\mismanaged-plastic-waste.csv')
```

```python
#Infos do dataset
print(data.head())
print(data.info())
print(data.describe())
print(data.columns)
```

```python
#alterações no dataset
data.rename(columns={'Entity': 'Country'}, inplace=True)
data.rename(columns={'Mismanaged': 'MismanagedPlastic'}, inplace=True)
```

```python
#visualizações dos dados
data['Country'].unique()

```

```python
regions = [
    "Americas (excl. USA)",
    "Asia (excl. China and India)",
    "Europe",
    "Middle East & North Africa",
    "Oceania",
    "Sub-Saharan Africa",
    "World"
]

# separar datasets
df_regions = data[data["Country"].isin(regions)].copy()
df_countries = data[~data["Country"].isin(regions)].copy()

# guardar CSVs
df_countries.to_csv("datasets\\mismanaged_plastic_waste_countries.csv", index=False)
df_regions.to_csv("datasets\\mismanaged_plastic_waste_regions.csv", index=False)

```
